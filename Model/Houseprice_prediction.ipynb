```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"T:\My Drive\000 ihsan\02 Courses\003GIT\MYProject\Housing_rate_regression.py_ML\housing_data.csv")
data.shape
```

```python
data.head(5)
```

```python
data.describe()
```

```python
x = data[["Square Footage",	"Number of Bedrooms",	"Number of Bathrooms",	"Lot Size (sq ft)",	"Year Built",	"Garage Spaces",	"Proximity to City Center (miles)",	"School District Rating",	"Crime Rate",	"Neighborhood Popularity Score",	"Energy Efficiency Rating"]]
x.head(3)
```

```python
y = data.iloc[:, 0]  # This selects the first column (Price) and all rows
print(y.head(3))  # Display the first 3 rows
```

```python
print(x.shape , y.shape)
y = np.array(y).reshape(-1, 1)
print(x.shape , y.shape)
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)# randomstate 42 - any time randome selection same , no change time to time
```

```python
print(x_train.shape , x_test.shape , y_train.shape , y_test.shape)
```

```python
model = linear_model.LinearRegression(fit_intercept=True)

```

```python
model.fit(x_train,y_train)
```

```python
y_predict = model.predict(x_test)
y_predict
```

```python
MSE = mean_squared_error(y_test,y_predict)
MSE
```

```python
# Line plot for training data
plt.plot(x_train, y_train, color="red", label="Train data (Red)", linestyle="--")

# Plot the regression line
plt.plot(x_test, y_predict, color="blue", label="Regression line", linewidth=2)

# Add title and labels
plt.title("Training Data and Regression Line")
plt.xlabel("X values")
plt.ylabel("Y values")

# Add a legend
plt.legend()

file_path = r"T:\My Drive\000 ihsan\02 Courses\003GIT\MYProject\Housing_rate_regression.py_ML\Plot\output2.jpg"
plt.savefig(file_path, format='jpg', dpi=300)

# Show the plot
plt.show()
```
